# Importando Bibliotecas

```
import numpy as np
import pandas as pd
import string
import warnings
warnings.filterwarnings('ignore')

import requests
```

# Obtendo e tratando os dados

```
url = 'http://www.fundamentus.com.br/resultado.php'
```

```
header = {
  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"
 }

r = requests.get(url, headers=header)
```

```
df = pd.read_html(r.text,  decimal=',', thousands='.')[0]
```

```
df['LPA'] = df['Cotação'] / df['P/L']
```

```
df
```

###Número de Empresas listadas

```
df.shape
```

```
for coluna in ['Div.Yield', 'Mrg Ebit', 'Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
  df[coluna] = df[coluna].str.replace('.', '')
  df[coluna] = df[coluna].str.replace(',', '.')
  df[coluna] = df[coluna].str.rstrip('%').astype('float') / 100
```

```
df
```

## Analisando os dados

#### - Buscando Small Caps - Buscar os indicadores corretos


```
df = df[df['Liq.2meses'] > 1000]
```

```
df.shape
```

```
ranking = pd.DataFrame()
ranking['pos'] = range(1,200)
ranking['EV/EBIT'] = df[ df['EV/EBIT'] > 0 ].sort_values(by=['EV/EBIT'])['Papel'][:199].values
ranking['ROIC'] = df.sort_values(by=['ROIC'], ascending=False)['Papel'][:199].values
ranking['Liq. Corr.'] = df[ df['Liq. Corr.'] >= 1.5 ].sort_values(by=['Liq. Corr.'])['Papel'][:199].values
ranking['ROE'] = df.sort_values(by=['ROE'], ascending=False)['Papel'][:199].values
```

```
ranking
```

```
a = ranking.pivot_table(columns='EV/EBIT', values='pos')
```

```
b = ranking.pivot_table(columns='ROIC', values='pos')
```

```
c = ranking.pivot_table(columns='Liq. Corr.', values='pos')
```

```
d = ranking.pivot_table(columns='ROE', values='pos')
```

```
t = pd.concat([a,b,c,d])
t
```

```
rank = t.dropna(axis=1).sum()
rank
```

```
rank.sort_values()[:20]
```

```
from tabulate import tabulate

# Filtrar empresas únicas
empresas_unicas = set()
resultado = []

for empresa in rank.sort_values()[:20].index:
    sigla = empresa[:4]  # Obtém as quatro primeiras letras da sigla
    if sigla not in empresas_unicas:
        empresas_unicas.add(sigla)
        resultado.append([empresa, rank[empresa]])

# Exibir o resultado como tabela
print(tabulate(resultado, headers=["Empresa", "Ranking"], tablefmt="pretty"))
```
